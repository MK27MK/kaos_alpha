## Stochastic processes

Here I'm going to define the stochastic processes that I'm going to use to generate synthetic prices throughout all the different levels of the game

## Level 1

Let's define our first and, most simple process.

This process will consist of a simple sin function with some noise added to it.

```python
NUM_OF_PRICES: int = 100
```

```python
import matplotlib.pyplot as plt

from processes import GlorifiedSin

sin_instrument = GlorifiedSin(noise_std=1.5)  # 1,5 produces good results
sin_prices = [sin_instrument.next_price() for _ in range(NUM_OF_PRICES)]
fig, ax = plt.subplots()

ax.plot(sin_prices, linewidth=2, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$")
ax.set(xlabel="time", ylabel="price")
plt.show()
```
